```python
import pygame
import time

pygame.init()  # pygame 사용전 초기화 (필수)

window_width = 689
window_height = 780
walkCount = 0
walkCount2 = 0
window_size = [window_width, window_height]  # 1280x640 로 화면 크기 설정
window = pygame.display.set_mode(window_size)  # 화면 크기 적용
pygame.display.set_caption("pygame 캐릭터 이동 코드")  # 창 이름
image_file = pygame.image.load("img/test_map.png") # 사진 불러오기
fps = pygame.time.Clock() 

up_image = [
    pygame.image.load("img/main_char/char_back1.png"),
    pygame.image.load("img/main_char/char_back2.png"),
    pygame.image.load("img/main_char/char_back3.png")
]
down_image = [
    pygame.image.load("img/main_char/char_front1.png"),
    pygame.image.load("img/main_char/char_front2.png"),
    pygame.image.load("img/main_char/char_front3.png")
]
left_image = [
    pygame.image.load("img/main_char/char_left1.png"),
    pygame.image.load("img/main_char/char_left2.png"),
    pygame.image.load("img/main_char/char_left3.png")
]
right_image = [
    pygame.image.load("img/main_char/char_right1.png"),
    pygame.image.load("img/main_char/char_right2.png"),
    pygame.image.load("img/main_char/char_right3.png")
]
character = pygame.image.load("img/main_char/char_front2.png")
character_size = character.get_rect().size # 캐릭터 이미지의 크기를 구한다.
character_width = character_size[0] # 캐릭터 이미지의 가로 크기
character_height = character_size[1] # 캐릭터 이미지의 세로 크기
character_x_pos = (window_width / 2) - (character_width / 2) # 화면 가로의 절반크기(중앙)에 해당하는 곳에 위치
character_y_pos = window_height - character_height # 화면 세로의 가장 아래에 해당하는 곳에 위치

# 이동할 좌표
to_x = 0
to_y = 0

is_left = False
is_right = False
is_up = False
is_down = False

def playerMove():
    global walkCount, character_x_pos, character_y_pos
    
    if walkCount > 2:
        walkCount = 0
        
    if is_left == True:
        window.blit(left_image[walkCount], (character_x_pos, character_y_pos))
        walkCount += 1
    elif is_right == True:
        window.blit(right_image[walkCount], (character_x_pos, character_y_pos))
        walkCount += 1
    elif is_up == True:
        window.blit(up_image[walkCount], (character_x_pos, character_y_pos))
        walkCount += 1
    elif is_down == True:
        window.blit(down_image[walkCount], (character_x_pos, character_y_pos))
        walkCount += 1
    else:
        window.blit(character, (character_x_pos, character_y_pos))

play = True
while play:
    deltatime = fps.tick(16) # fps 지정해주지 않으면 원이 너무 빨리 움직임
    for event in pygame.event.get():
        if event.type == pygame.QUIT: #game의 event type이 QUIT 명령이라면
            play = False # while문을 종료시킴
        if event.type == pygame.KEYDOWN: #game의 event type이 키보드를 누른 상태일 때
            if event.key == pygame.K_w:
                to_y = -5 # 좌표를 직접 변화시키지 않고 변수에 저장해 두었다가
                is_up = True
            elif event.key == pygame.K_s:
                to_y = 5
                is_down = True
            elif event.key == pygame.K_d:
                to_x = 5
                is_right = True
            elif event.key == pygame.K_a:
                to_x = -5
                is_left = True
            elif event.key == pygame.K_q:
                to_x = -2.5
                to_y = -2.5
                is_up = True
            elif event.key == pygame.K_e:
                to_x = 2.5
                to_y = -2.5
                is_up = True
            elif event.key == pygame.K_z:
                to_x = -2.5
                to_y = 2.5
                is_down = True
            elif event.key == pygame.K_c:
                to_x = 2.5
                to_y = 2.5
                is_down = True
        if event.type == pygame.KEYUP:
            # 키 별로 변수값을 나눠 두 개의 방향키를 동시에 눌러도 문제 없도록 함
            if event.key == pygame.K_w:
                to_y = 0
                is_up = False
            elif event.key == pygame.K_s:
                to_y = 0
                is_down = False
            elif event.key == pygame.K_d:
                to_x = 0
                is_right = False
            elif event.key == pygame.K_a:
                to_x = 0
                is_left = False
            elif event.key == pygame.K_q:
                to_x = 0
                to_y = 0
                is_up = False
                is_left = False
            elif event.key == pygame.K_e:
                to_x = 0
                to_y = 0
                is_up = False
                is_right = False
            elif event.key == pygame.K_z:
                to_x = 0
                to_y = 0
                is_down = False
                is_left = False
            elif event.key == pygame.K_c:
                to_x = 0
                to_y = 0
                is_down = False
                is_right = False
                
    window.blit(image_file, (0, 0)) # 사진 그리기 
    playerMove()
    character_x_pos += to_x # 캐릭터의 x좌표 위치 계산
    character_y_pos += to_y # 캐릭터의 y좌표 위치 계산
    #window.blit(character, (character_x_pos, character_y_pos)) # 캐릭터 그리기
    
    pygame.display.update() # 화면 업데이트

pygame.quit()  # pygame 종료
```
