```python
import pygame
import sys
import random
import math
from utils import scale_image, blit_rotate_center

BLACK = (0, 0, 0)

MAZE_W = 11
MAZE_H = 9
maze = []
for y in range(MAZE_H):
    maze.append([0] * MAZE_W)

DUNGEON_W = MAZE_W * 3
DUNGEON_H = MAZE_H * 3
dungeon = []
for y in range(DUNGEON_H):
    dungeon.append([0] * DUNGEON_W)

imgWall = pygame.image.load("img/wall.png")
imgFloor = pygame.image.load("img/floor.png")
imgPlayer = pygame.image.load("img/player.png")



def make_dungeon():
    XP = [0, 1, 0, -1]
    YP = [-1, 0, 1, 0]
    
    #주변 벽
    for x in range(MAZE_W):
        maze[0][x] = 1
        maze[MAZE_H - 1][x] = 1
    for y in range(1, MAZE_H - 1):
        maze[y][0] = 1
        maze[y][MAZE_W - 1] = 1
    
    #내부 비우기
    for y in range(2, MAZE_H - 1):
        for x in range(1, MAZE_W - 1):
            maze[y][x] = 0
    
    #기둥
    for y in range(2, MAZE_H - 2, 2):
        for x in range(2, MAZE_W - 2, 2):
            maze[y][x] = 1
            
    #기둥에서 상하좌우로 벽 생성
    for y in range(2, MAZE_H - 2, 2):
        for x in range(2, MAZE_W - 2, 2):
            d = random.randint(0, 3)
            if x > 2:
                d = random.randint(0, 2)
            maze[y + YP[d]][x + XP[d]] = 1

    # 미로에서 던전 생성
    # 전체를 벽으로
    for y in range(DUNGEON_H):
        for x in range(DUNGEON_W):
            dungeon[y][x] = 9

    # 방과 미로 배치
    for y in range(1, MAZE_H - 1):
        for x in range(1, MAZE_W - 1):
            dx = x * 3 + 1
            dy = y * 3 + 1
            if maze[y][x] == 0:
                if random.randint(0, 99) < 20:
                    for ry in range(-1, 2):
                        for rx in range(-1, 2):
                            dungeon[dy + ry][dx + rx] = 0
                # 미로 생성
                else:
                    dungeon[dy][dx] = 0
                    if maze[y - 1][x] == 0:
                        dungeon[dy - 1][dx] = 0
                    if maze[y + 1][x] == 0:
                        dungeon[dy + 1][dx] = 0
                    if maze[y][x - 1] == 0:
                        dungeon[dy][dx - 1] = 0
                    if maze[y][x + 1] == 0:
                        dungeon[dy][dx + 1] = 0

# 던전 그리기
def draw_dungeon(bg):
    bg.fill(BLACK)
    for y in range(-5, 40):
        for x in range(-5, 40):
            X = (x + 5) * 20
            Y = (y + 5) * 20
            dx = 5 + x
            dy = 5 + y
            if 0 <= dx and dx < DUNGEON_W and 0 <= dy and dy < DUNGEON_H:
                if dungeon[dy][dx] == 0:
                    bg.blit(imgFloor, [X, Y])
                if dungeon[dy][dx] == 9:
                    bg.blit(imgWall, [X , Y])
            if x == 0 and y == 0:
                bg.blit(imgPlayer, [X, Y - 8])
"""
class CameraGroup(pygame.sprite.Group):
    def __init__(self):
        super().__init__()
        self.display_surface = pygame.display.get_surface()

        # camera offset 
        self.offset = pygame.math.Vector2()
        self.half_w = self.display_surface.get_size()[0] // 2
        self.half_h = self.display_surface.get_size()[1] // 2

        # ground
        self.ground_surf = make_dungeon()
        self.ground_rect = self.ground_surf.get_rect(topleft = (0,0))
        
    def center_target_camera(self, target):
        self.offset.x = self.target.rect.centerx - self.half_w
        self.offset.y = self.target.rect.centery - self.half_h

    def custom_draw(self,player):
        # ground 
        ground_offset = self.ground_rect.topleft - self.offset + self.internal_offset
        self.internal_surf.blit(self.ground_surf,ground_offset)

        # active elements
        for sprite in sorted(self.sprites(),key = lambda sprite: sprite.rect.centery):
            offset_pos = sprite.rect.topleft - self.offset + self.internal_offset
            self.internal_surf.blit(sprite.image,offset_pos)
"""
class AbstractCar:
    
    def __init__(self, max_vel, rotation_vel):
        self.img = self.IMG
        self.max_vel = max_vel
        self.vel = 0
        self.rotation_vel = rotation_vel
        self.angle = 0
        self.x, self.y = self.START_POS
        self.acceleration = 0.1
        
    def rotate(self, left=False, right=False):
        if left:
            self.vel = self.max_vel
            self.move_horizontal()
        elif right:
            self.vel = -self.max_vel
            self.move_horizontal()
    
    def draw(self, win):
        blit_rotate_center(win, self.img, (self.x, self.y), self.angle)
        
    def move_forward(self):
        self.vel = self.max_vel
        self.move_vertical()
        
    def move_backward(self):
        self.vel = -self.max_vel
        self.move_vertical()
        
    def move_vertical(self):
        radians = math.radians(self.angle)
        vertical = math.cos(radians) * self.vel
        self.y -= vertical
        
    def move_horizontal(self):
        radians = math.radians(self.angle)
        horizontal = math.cos(radians) * self.vel
        self.x -= horizontal
        
class PlayerCar(AbstractCar):
    IMG = imgPlayer
    START_POS = (50, 50)
    
    def reduce_speed(self):
        self.vel = 0
        self.move_vertical()
        self.move_horizontal()
        
def move_player(player_car):
    keys = pygame.key.get_pressed()
    moved = False

    if keys [pygame.K_a]:
        player_car.rotate(left = True)
    if keys [pygame.K_d]:
        player_car.rotate(right = True)
    if keys[pygame.K_w]:
        moved = True
        player_car.move_forward()
    if keys[pygame.K_s]:
        moved = True
        player_car.move_backward()
        # s를 누르면 player_car가 후진
        
    if not moved:
        player_car.reduce_speed()
        
def draw(win, player_car):
    player_car.draw(win)
    pygame.display.update()
    
def collide(self, mask, x=0, y=0):
    car_mask = pygame.mask.from_surface(self.img) # self.img의 마스크를 생성
    offset = (int(self.x - x), int(self.y - y)) # self의 현재 위치와 특정 위치 사이의 상대적인 거리를 정수로 표현
    poi = mask.overlap(car_mask, offset) # mask와 car_mask를 겹쳐서 충돌 여부 판단
    return poi # 충돌 여부 리턴

def draw_text(bg, txt, x, y, fnt, col): # 그림자 포함한 문자 표시
    sur = fnt.render(txt, True, BLACK)
    bg.blit(sur, [x + 1, y + 2])
    sur = fnt.render(txt, True, col)
    bg.blit(sur, [x, y])
    
def main():
    pygame.init()
    pygame.display.set_caption("던전 내 걷기")
    screen = pygame.display.set_mode((880, 720))
    clock = pygame.time.Clock()
    player_car = PlayerCar(5, 5)
    WHITE = (255, 255, 255)
    floor = 0
    fontS = pygame.font.Font(None, 30)
    # camera_group = CameraGroup()
    
    make_dungeon()
    
    while True:
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
                
        
        move_player(player_car)
        draw_dungeon(screen)
        #draw_text(screen, "floor {} ({},{})".format(floor, pl_x, pl_y), 60, 40, fontS, WHITE)
        draw(screen, player_car)
        
        #camera_group.update()
        #camera_group.custom_draw()
        pygame.display.update()
        clock.tick(60)

if __name__ == '__main__':
    main()
```
