```python
import pygame
import math
import sys
import random
import speech_recognition as sr
import numpy as np
import time
import threading
import moviepy.editor as mp
from pygame.locals import *
from weather import *

# 색 정의
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
CYAN = (0, 255, 255)
BLINK = [(224, 255, 255), (192, 240, 255), (128, 224, 255), (64, 192, 255), (128, 224, 255), (192, 240, 255)]

screen = pygame.display.set_mode((880, 720))
canvas = pygame.Surface((880, 720))
video = mp.VideoFileClip("video/intro.mp4")

# 이미지 로딩
imgTitle = pygame.image.load("img/title.png")
imgWall = pygame.image.load("img/wall.png")
imgWall2 = pygame.image.load("img/wall2.png")
imgDark = pygame.image.load("img/dark.png")
imgSnow = pygame.image.load('img/snow_.png')
imgSnow2 = pygame.image.load('img/snow2.png').convert_alpha()
imgRain = pygame.image.load('img/rain_.png')
imgRain2 = pygame.image.load('img/rain2.png').convert_alpha()
imgCloud = pygame.image.load('img/cloud.png').convert_alpha()
imgPara = pygame.image.load("img/parameter.png")
imgBtlBG = pygame.image.load("img/btlbg.png")
imgEnemy = pygame.image.load("img/enemy0.png")
imgItem =  [
    pygame.image.load("img/potion.png"),
    pygame.image.load("img/spoiled.png"),
    pygame.image.load("img/fist.png"),
    pygame.image.load("img/spear.png"),
    pygame.image.load("img/pistol.png"),
    pygame.image.load("img/apple.png"),
    pygame.image.load("img/meat.png")
]
imgItemIcon = [
    pygame.image.load("img/fist_icon.png"),
    pygame.image.load("img/spear_icon.png"),
    pygame.image.load("img/pistol_icon.png")
]
imgFloor = [
    pygame.image.load("img/floor.png"),
    pygame.image.load("img/tbox.png"),
    pygame.image.load("img/cocoon.png"),
    pygame.image.load("img/stairs.png")
]
imgPlayer = [
    pygame.image.load("img/mychr0.png"),
    pygame.image.load("img/mychr1.png"),
    pygame.image.load("img/mychr2.png"),
    pygame.image.load("img/mychr3.png"),
    pygame.image.load("img/mychr4.png"),
    pygame.image.load("img/mychr5.png"),
    pygame.image.load("img/mychr6.png"),
    pygame.image.load("img/mychr7.png"),
    pygame.image.load("img/mychr8.png")
]
imgEffect = [
    pygame.image.load("img/effect_a.png"),
    pygame.image.load("img/effect_b.png")
]

# 변수 선언
i2 = 0
speed = 1
idx = 25
tmr = 0
floor = 0
fl_max = 1
welcome = 0

pl_x = 0
pl_y = 0
pl_d = 0
pl_a = 0
pl_lifemax = 0
pl_life = 0
pl_str = 0
food = 0
potion = 0
attack_type = 0
treasure = 0
type_system = 0

emy_name = ""
emy_lifemax = 0
emy_life = 0
emy_str = 0
emy_x = 0
emy_y = 0
emy_step = 0
emy_blink = 0
dmg_eff = 0
btl_cmd = 0
alpha = 255

COMMAND = ["[O]rder"]
TRE_NAME = ["Potion", "Food spoiled", "  Fist", "Spear", "Pistol", "Food +20", "Food +100"]
EMY_NAME = [
    "Green slime", "Red slime", "Axe beast", "Ogre", "Sword man", "Death hornet", "Signal slime", "Devil plant", "Twin killer", "Hell"
]

MAZE_W = 11
MAZE_H = 9
maze = []
for y in range(MAZE_H):
    maze.append([0] * MAZE_W)

DUNGEON_W = MAZE_W * 5
DUNGEON_H = MAZE_H * 5
dungeon = []
for y in range(DUNGEON_H):
    dungeon.append([0] * DUNGEON_W)

# 던전 자동 생성
def make_dungeon():
    XP = [0, 1, 0, -1]
    YP = [-1, 0, 1, 0]
    
    # 테두리 벽
    for x in range(MAZE_W):
        maze[0][x] = 1
        maze[MAZE_H - 1][x] = 1
    for y in range(1, MAZE_H - 1):
        maze[y][0] = 1
        maze[y][MAZE_W - 1] = 1
    
    # 가운데 아무것도 없는 상태
    for y in range(1, MAZE_H - 1):
        for x in range(1, MAZE_W - 1):
            maze[y][x] = 0
    
    #기둥
    for y in range(2, MAZE_H - 2, 2):
        for x in range(2, MAZE_W - 2, 2):
            maze[y][x] = 1
    
    # 기둥에서 상하좌우로 벽 생성
    for y in range(2, MAZE_H - 2, 2):
        for x in range(2, MAZE_W - 2, 2):
            d = random.randint(0, 3)
            # 2열부터 왼쪽에는 벽 생성하지않음
            if x > 2:
                d = random.randint(0, 2)
            maze[y + YP[d]][x + XP[d]] = 1
    
    # 미로에서 던전 생성
    # 전체를 벽으로
    for y in range(DUNGEON_H):
        for x in range(DUNGEON_W):
            dungeon[y][x] = 9
    
    # 방과 통로 배치
    for y in range(1, MAZE_H - 1):
        for x in range(1, MAZE_W - 1):
            dx = x * 3 + 1
            dy = y * 3 + 1
            if maze[y][x] == 0:
                if random.randint(0, 99) < 20: # 방 생성
                    for ry in range(-1, 2):
                        for rx in range(-1, 2):
                            dungeon[dy + ry][dx + rx] = 0
                else: # 통로 생성
                    dungeon[dy][dx] = 0
                    if maze[y - 1][x] == 0:
                        dungeon[dy - 1][dx] = 0
                    if maze[y + 1][x] == 0:
                        dungeon[dy + 1][dx] = 0
                    if maze[y][x - 1] == 0:
                        dungeon[dy][dx - 1] = 0
                    if maze[y][x + 1] == 0:
                        dungeon[dy][dx + 1] = 0
                    
#던전 표시
def draw_dungeon(bg, fnt):
    bg.fill(BLACK)
    for y in range(-5, 6):
        for x in range(-5, 6):
            X = (x + 5) * 80
            Y = (y + 5) * 80
            dx = pl_x + x
            dy = pl_y + y
            if 0 <= dx and dx < DUNGEON_W and 0 <= dy and dy < DUNGEON_H:
                if dungeon[dy][dx] <= 3:
                    bg.blit(imgFloor[dungeon[dy][dx]], [X, Y])
                if dungeon[dy][dx] == 9:
                    bg.blit(imgWall, [X, Y - 40])
                    if dy >= 1 and dungeon[dy - 1][dx] == 9:
                        bg.blit(imgWall2, [X, Y - 80])
            if x == 0 and y == 0: # 주인공 캐릭터 표시
                bg.blit(imgPlayer[pl_a], [X, Y - 40])
    bg.blit(imgDark, [0, 0])
    draw_para(bg, fnt) # 주인공 능력 표시

# 길에 이벤트 배치
def put_event(): 
    global pl_x, pl_y, pl_d, pl_a
    
    # 계단 배치
    while True:
        x = random.randint(3, DUNGEON_W - 4)
        y = random.randint(3, DUNGEON_H - 4)
        if (dungeon[y][x] == 0):
            for ry in range(-1, 2): # 계단 주변을 길로 만듦
                for rx in range(-1, 2):
                    dungeon[y + ry][x + rx] = 0
            dungeon[y][x] = 3
            break
    
    # 보물 상자와 누에고치 배치
    for i in range(60):
        x = random.randint(3, DUNGEON_W - 4)
        y = random.randint(3, DUNGEON_H - 4)
        if (dungeon[y][x] == 0):
            dungeon[y][x] = random.choice([1, 2, 2, 2, 2])
    
    # 플레이어 초기 위치
    while True:
        pl_x = random.randint(3, DUNGEON_W - 4)
        pl_y = random.randint(3, DUNGEON_H - 4)
        if (dungeon[pl_y][pl_x] == 0):
            break
    pl_d = 1
    pl_a = 2
    
# 주인공 이동
def move_player(key):
    global idx, tmr, pl_x, pl_y, pl_d, pl_a, pl_life, food, potion, blazegem, treasure, attack_type
    
    # 보물 상자에 닿음
    if dungeon[pl_y][pl_x] == 1: 
        dungeon[pl_y][pl_x] = 0
        treasure = random.choice([2, 3, 4])
        if treasure == 0:
            potion += 1
        if treasure == 1:
            food = int(food / 2)
        if treasure == 2:
            attack_type = 0
        if treasure == 3:
            attack_type = 1
        if treasure == 4:
            attack_type = 2
        idx = 3
        tmr = 0
        return
    
    # 누에고치에 닿음
    if dungeon[pl_y][pl_x] == 2: 
        dungeon[pl_y][pl_x] = 0
        r = random.randint(0, 99)
        if floor <= 0:
            if r < 60:
                treasure = random.choice([5, 5, 5, 6])
                if treasure == 5: food += 20
                elif treasure == 6: food += 100
                idx = 3
                tmr = 0
            else: # 적 출현
                idx = 10
                tmr = 0
        if floor >= 3:
            if r < 50:
                treasure = random.choice([5, 5, 5, 6])
                if treasure == 5: food += 20
                elif treasure == 6: food += 100
                idx = 3
                tmr = 0
            else: # 적 출현
                idx = 10
                tmr = 0
        if floor >= 5:
            if r < 40:
                treasure = random.choice([5, 5, 5, 6])
                if treasure == 5: food += 20
                elif treasure == 6: food += 80
                idx = 3
                tmr = 0
            else: # 적 출현
                idx = 10
                tmr = 0
        if floor >= 7:
            if r < 30:
                treasure = random.choice([5, 5, 5, 6])
                if treasure == 5: food += 10
                elif treasure == 6: food += 60
                idx = 3
                tmr = 0
            else: # 적 출현
                idx = 10
                tmr = 0
        else:
            if r < 20:
                treasure = random.choice([5, 5, 5, 6])
                if treasure == 5: food += 5
                elif treasure == 6: food += 40
                idx = 3
                tmr = 0
            else: # 적 출현
                idx = 10
                tmr = 0
        return
    
    # 계단에 닿음
    if dungeon[pl_y][pl_x] == 3:
        idx = 2
        tmr = 0
        return
    
    # 방향 키로 상하좌우 이동
    x = pl_x
    y = pl_y
    for event in pygame.event.get():  # 이벤트 발생 여부 확인
        if event.type == pygame.QUIT:  # 창이 닫히는 이벤트 발생시
            is_game_running = False  # 게임 진행 여부 -> False

        if event.type == pygame.KEYDOWN: # 키가 눌러졌는지 확인
            if event.key == pygame.K_LEFT: # 위쪽 화살표 키를 눌렀을때
                pl_x -= 2 # 왼쪽으로 5만큼 위치이동
            if event.key == pygame.K_RIGHT: # 아래쪽 화살표 키를 눌렀을때
                pl_x += 2  # 오른쪽으로 5만큼 위치이동
            if event.key == pygame.K_UP: # 오른쪽 화살표 키를 눌렀을때
                pl_y -= 2 # 위로 5만큼 위치이동
            if event.key == pygame.K_DOWN: # 왼쪽 화살표 키를 눌렀을때
                pl_y += 2  # 아래로 5만큼 위치이동

        if event.type == pygame.KEYUP: # 키를 떼엇을때 멈춘다
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                pl_x = 0
            elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                pl_y = 0
    
    char_x = pl_x
    char_y = pl_y
    
    if key == pygame.KEYUP: # 키를 떼엇을때 멈춘다
        if key == pygame.K_LEFT or key == pygame.K_RIGHT:
            pl_x = 0
        elif key == pygame.K_UP or key == pygame.K_DOWN:
            pl_y = 0
        
    pl_a = pl_d * 2
    if pl_x != x or pl_y != y: # 이동 시 식량 및 체력 계산
        pl_a += tmr % 2 # 이동 시 걷는 애니메이션
        if food > 0:
            food -= 1
            if pl_life < pl_lifemax:
                pl_life += 1
        else:
            pl_life -= 5
            if pl_life <= 0:
                pl_life = 0
                pygame.mixer.music.stop()
                idx = 9
                tmr = 0

# 그림자 포함한 문자 표시
def draw_text(bg, txt, x, y, fnt, col): 
    sur = fnt.render(txt, True, BLACK)
    bg.blit(sur, [x + 1, y + 2])
    sur = fnt.render(txt, True, col)
    bg.blit(sur, [x, y])
    
# 주인공 능력 표시
def draw_para(bg, fnt):
    global attack_type
    X = 30
    Y = 600
    type = ['Fist', 'Spear', 'Pistol']
    bg.blit(imgPara, [X, Y])
    col = WHITE
    
    if pl_life < 10 and tmr % 2 == 0: col = RED
    draw_text(bg, "{}/{}".format(pl_life, pl_lifemax), X + 128, Y + 6, fnt, col)
    draw_text(bg, str(pl_str), X + 128, Y + 33, fnt, WHITE)
    col = WHITE
    if food == 0 and tmr % 2 == 0: col = RED
    draw_text(bg, str(food), X + 128, Y + 60, fnt, col)
    draw_text(bg, str(potion), X + 266, Y + 6, fnt, WHITE)
    draw_text(bg, "{}".format(type[attack_type]), X + 235, Y + 60, fnt, WHITE)
    bg.blit(imgItemIcon[attack_type], [X + 210, Y + 60])

# 전투 시작 준비
def init_battle(): 
    global imgEnemy, emy_name, emy_lifemax, emy_life, emy_str, emy_x, emy_y, type_num, emy_type
    typ = random.randint(0, floor)
    
    if floor >= 10:
        typ = random.randint(0, 9)
    lev = random.randint(1, floor)
    imgEnemy = pygame.image.load("img/enemy" + str(typ) + ".png")
    emy_name = EMY_NAME[typ] + " LV" + str(lev)
    emy_lifemax = 60 * (typ + 1) + (lev - 1) * 10
    emy_life = emy_lifemax
    emy_str = int(emy_lifemax / 8)
    emy_x = 440 - imgEnemy.get_width() / 2
    emy_y = 560 - imgEnemy.get_height()
    type_num = random.randint(0, 2)
    if type_num == 0:
        emy_type = pygame.image.load("img/fist.png")
    elif type_num == 1:
        emy_type = pygame.image.load("img/spear.png")
    elif type_num == 2:
        emy_type = pygame.image.load("img/pistol.png")

# 적 체력 표시 게이지
def draw_bar(bg, x, y, w, h, val, ma):
    pygame.draw.rect(bg, WHITE, [x - 2, y - 2, w + 4, h + 4])
    pygame.draw.rect(bg, BLACK, [x, y, w, h])
    if val > 0:
        pygame.draw.rect(bg, (0, 128, 255), [x, y, w * val / ma, h])

# 전투 화면 표시
def draw_battle(bg, fnt):
    global emy_blink, dmg_eff, type_system
    bx = 0
    by = 0
    X = 30
    Y = 600
    
    if dmg_eff > 0:
        dmg_eff -= 1
        bx = random.randint(-20, 20)
        by = random.randint(-10, 10)
    bg.blit(imgBtlBG, [bx, by])
    if emy_life > 0 and emy_blink % 2 == 0:
        bg.blit(imgEnemy, [emy_x, emy_y + emy_step])
    draw_bar(bg, 340, 580, 200, 10, emy_life, emy_lifemax)
    bg.blit(emy_type, [550, 560])
    if attack_type == 0 and type_num == 0: # fist vs fist == Normal
        draw_text(bg, "Normal", X + 310, Y + 62, fnt, WHITE)
        type_system = 0
    elif attack_type == 0 and type_num == 1: #fist vs spear == Good
        draw_text(bg, "Good", X + 310, Y + 62, fnt, WHITE)
        type_system = 1
    elif attack_type == 0 and type_num == 2: # fist vs pistol == Bad
        draw_text(bg, "Bad", X + 310, Y + 62, fnt, WHITE)
        type_system = 2
    elif attack_type == 1 and type_num == 0: # spear vs fist == Bad
        draw_text(bg, "Bad", X + 310, Y + 62, fnt, WHITE)
        type_system = 2
    elif attack_type == 1 and type_num == 1: # spear vs spear == Normal
        draw_text(bg, "Normal", X + 310, Y + 62, fnt, WHITE)
        type_system = 0
    elif attack_type == 1 and type_num == 2: # spear vs pistol == Good
        draw_text(bg, "Good", X + 310, Y + 62, fnt, WHITE)
        type_system = 1
    elif attack_type == 2 and type_num == 0: # pistol vs fist == Good
        draw_text(bg, "Good", X + 310, Y + 62, fnt, WHITE)
        type_system = 1
    elif attack_type == 2 and type_num == 1: # pistol vs spear == Bad
        draw_text(bg, "Bad", X + 310, Y + 62, fnt, WHITE)
        type_system = 2
    elif attack_type == 2 and type_num == 2: #pistol vs pistol == Normal
        draw_text(bg, "Normal", X + 310, Y + 62, fnt, WHITE)
        type_system = 0
    
    if emy_blink > 0:
        emy_blink -= 1
    for i in range(10): # 전투 메세지 표시
        draw_text(bg, message[i], 600, 100 + i * 50, fnt, WHITE)
        draw_para(bg, fnt) # 주인공 능력 표시

# 커맨드 입력 및 표시
def battle_command(bg, fnt, key):
    global btl_cmd, vToText
    ent = False
    btl_cmd = 0
    if key[K_o]: # O키
        btl_cmd = 1
        ent = True
    if key[K_SPACE] or key[K_RETURN]:
        ent = True
    c = WHITE
    if btl_cmd == 0: c = BLINK[tmr % 6]
    draw_text(bg, COMMAND[0], 20, 360 + 0 * 60, fnt, c)
    return ent

#전투 메세지 표시 처리
message = [""] * 10
def init_message():
    for i in range(10):
        message[i] = ""

def set_message(msg):
    for i in range(10):
        if message[i] == "":
            message[i] = msg
            return
    for i in range(9):
        message[i] = message[i + 1]
    message[9] = msg

# 메인 처리
def main():
    global speed, idx, tmr, floor, fl_max, welcome, FPS, num
    global pl_a, pl_lifemax, pl_life, pl_str, food, potion, attack_type, pl_x, pl_y
    global emy_life, emy_step, emy_blink, dmg_eff
    dmg = 0
    lif_p = 0
    str_p = 0
    
    pygame.init()
    pygame.display.set_caption("One Hour Dungeon")
    clock = pygame.time.Clock()
    font = pygame.font.Font(None, 40)
    fontS = pygame.font.Font(None, 30)
    weather = proc_weather().split('.')
    Y = [125, 150, 175]
    
    se = [ # 효과음 및 징글
        pygame.mixer.Sound("sound/ohd_se_attack.ogg"),
        pygame.mixer.Sound("sound/ohd_se_blaze.ogg"),
        pygame.mixer.Sound("sound/ohd_se_potion.ogg"),
        pygame.mixer.Sound("sound/ohd_jin_gameover.ogg"),
        pygame.mixer.Sound("sound/ohd_jin_levup.ogg"),
        pygame.mixer.Sound("sound/ohd_jin_win.ogg")
    ]

    while True:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            if event.type == KEYDOWN:
                if event.key == pygame.K_LEFT: # 위쪽 화살표 키를 눌렀을때
                    pl_x -= 1 # 왼쪽으로 5만큼 위치이동
                if event.key == pygame.K_RIGHT: # 아래쪽 화살표 키를 눌렀을때
                    pl_x += 1  # 오른쪽으로 5만큼 위치이동
                if event.key == pygame.K_UP: # 오른쪽 화살표 키를 눌렀을때
                    pl_y -= 1 # 위로 5만큼 위치이동
                if event.key == pygame.K_DOWN: # 왼쪽 화살표 키를 눌렀을때
                    pl_y += 1  # 아래로 5만큼 위치이동
                        
        #canvas.fill(BLACK)
        #screen.blit(canvas, (0, 0, 0, 0))
        tmr += 1
        key = pygame.key.get_pressed()
        
        # 타이틀 화면
        if idx == 0:
            if tmr == 1:
                pygame.mixer.music.load("sound/ohd_bgm_title.ogg")
                pygame.mixer.music.play(-1)
            screen.fill(BLACK)
            screen.blit(imgTitle, [40, 60])
            if fl_max >= 2:
                draw_text(screen, "You reached floor {}.".format(fl_max), 300, 460, font, CYAN)
            draw_text(screen, "Press space key", 320, 560, font, BLINK[tmr % 6])
            if key[K_SPACE] == 1:
                make_dungeon()
                put_event()
                floor = 1
                welcome = 15
                pl_lifemax = 300
                pl_life = pl_lifemax
                pl_str  = 100
                food = 300
                potion = 0
                blazegem = 0
                idx = 1
                tmr = 0
                pygame.mixer.music.load("sound/ohd_bgm_field.ogg")
                pygame.mixer.music.play(-1)

        # 플레이어 이동
        elif idx == 1:
            move_player(event)
            draw_dungeon(screen, fontS)
            draw_text(screen, "floor {} ({},{})".format(floor, pl_x, pl_y), 60, 40, fontS, WHITE)
            draw_text(screen, "{}".format(tmr), 60, 80, fontS, WHITE)
            for i in range(3):
                t = threading.Thread(target=draw_text, args=(screen, "{}".format(weather[i]), 60, Y[i], fontS, WHITE))
                t.start()
            weather_filter(screen, fontS, 1)
            if welcome > 0:
                welcome -= 1
                draw_text(screen, "Welcome to floor {}.".format(floor), 300, 180, font, CYAN)
            if key[K_t] == 1:
                idx = 23
                tmr = 0
        
        # 다음 층 화면 전환
        elif idx == 2:
            draw_dungeon(screen, fontS)
            weather_filter(screen, fontS, 1)
            if 1 <= tmr and tmr <= 5:
                h = 80 * tmr
                pygame.draw.rect(screen, BLACK, [0, 0, 880, h])
                pygame.draw.rect(screen, BLACK, [0, 720 - h, 880, h])
            if tmr == 5:
                floor += 1
                if floor >= 11: # 엔딩으로
                    idx = 24
                    tmr = 0
                if floor > fl_max:
                    fl_max = floor
                welcome = 15
                make_dungeon()
                put_event()
            if tmr == 6:
                weather = proc_weather().split('.')
            if 7 <= tmr and tmr <= 10:
                h = 80 * (10 - tmr)
                pygame.draw.rect(screen, BLACK, [0, 0, 880, h])
                pygame.draw.rect(screen, BLACK, [0, 720 - h, 880, h])
            if tmr == 10:
                idx = 1
                tmr = 0
                
        # 아이템 입수 혹은 함정
        elif idx == 3:
            draw_dungeon(screen, fontS)
            weather_filter(screen, fontS, 1)
            screen.blit(imgItem[treasure], [340, 280])
            draw_text(screen, TRE_NAME[treasure], 400, 300, font, WHITE)
            draw_text(screen, "floor {} ({},{})".format(floor, pl_x, pl_y), 60, 40, fontS, WHITE)
            for i in range(3):
                t = threading.Thread(target=draw_text, args=(screen, "{}".format(weather[i]), 60, Y[i], fontS, WHITE))
                t.start()
                
            if tmr == 10:
                idx = 1
                tmr = 11
                
        # 게임 오버
        elif idx == 9:
            if tmr <= 30:
                PL_TURN = [2, 4, 0, 6]
                pl_a = PL_TURN[tmr % 4]
                if tmr == 30: pl_a = 8 # 쓰러진 이미지
                draw_dungeon(screen, fontS)
                weather_filter(screen, fontS, 1)
            elif 31 <= tmr and tmr <= 41:
                se[3].play()
                draw_text(screen, "You Die", 360, 240, font, RED)
                draw_text(screen, "Game over", 360, 380, font, RED)
            elif tmr >= 42:
                draw_text(screen, "Press R to try again", 320, 560, font, RED)
                if key[K_r] == 1:
                    idx = 0
                    tmr = 0
                
        # 전투 시작
        elif idx == 10:
            if tmr == 1:
                pygame.mixer.music.load("sound/ohd_bgm_battle.ogg")
                pygame.mixer.music.play(-1)
                init_battle()
                init_message()
            elif tmr <= 4:
                bx = (4 - tmr) * 220
                by = 0
                screen.blit(imgBtlBG, [bx, by])
                draw_text(screen, "Encounter!", 350, 200, font, WHITE)
            elif tmr <= 16:
                draw_battle(screen, fontS)
                draw_text(screen, emy_name + " appear!", 300, 200, font, WHITE)
            else:
                idx = 11
                tmr = 0
                
        # 플레이어 턴(입력 대기)
        elif idx == 11:
            draw_battle(screen, fontS)
            if tmr == 1: set_message("Your turn")
            if battle_command(screen, font, key) == True:
                if btl_cmd == 1:
                    try:
                        r = sr.Recognizer()
                        
                        with sr.Microphone() as source:
                            speech = r.listen(source, timeout=3, phrase_time_limit=2)
                            vToText = r.recognize_google(speech, language="ko-KR")
                            
                        if vToText == '공격':
                            idx = 12
                            tmr = 0
                        if vToText == '회복' and potion > 0:
                            idx = 20
                            tmr = 0
                        if vToText == '마법' and blazegem > 0:
                            idx = 21
                            tmr = 0
                        if vToText == '도망':
                            idx = 14
                            tmr = 0
                    except sr.UnknownValueError:
                        set_message("Try Again")
                        tmr = 2
                    except sr.RequestError as e:
                        set_message("Try Again")
                        tmr = 2
                    except sr.WaitTimeoutError as w:
                        set_message("Try Again")
                        tmr = 2
                    
        # 플레이어 공격
        elif idx == 12:
            draw_battle(screen, fontS)
            if tmr == 1:
                set_message("You attack")
                se[0].play()
                if type_system == 0:
                    dmg = pl_str + random.randint(0, 9)
                elif type_system == 1:
                    dmg = (pl_str + random.randint(0, 9) + 20) # 상성이 유리하면 dmg + 20
                elif type_system == 2:
                    dmg = (pl_str + random.randint(0, 9) - 20) # 상성이 불리하면 dmg - 20
            if 2 <= tmr and tmr <= 4:
                screen.blit(imgEffect[0], [700 - tmr * 120, -100 + tmr * 120])
            if tmr == 5:
                emy_blink = 5
                set_message(str(dmg) + "pts of damage!")
            if tmr == 11:
                emy_life -= dmg
                if emy_life <= 0:
                    emy_life = 0
                    idx = 16
                    tmr = 0
            if tmr == 16:
                idx = 13
                tmr = 0
                
        # 적 턴, 적 공격
        elif idx == 13:
            draw_battle(screen, fontS)
            if tmr == 1:
                set_message("Enemy turn")
            if tmr == 5:
                set_message(emy_name + " attack!")
                se[0].play()
                emy_step = 30
            if tmr == 9:
                dmg = emy_str + random.randint(0, 9)
                set_message(str(dmg) + "pts of damage!")
                dmg_eff = 5
                emy_step = 0
            if tmr == 15:
                pl_life -= dmg
                if pl_life < 0:
                    pl_life = 0
                    idx = 15
                    tmr = 0
            if tmr == 20:
                idx = 11
                tmr = 0
                
        # 후퇴 가능한가?
        elif idx == 14:
            draw_battle(screen, fontS)
            if tmr == 1: set_message("...")
            if tmr == 2: set_message(".....")
            if tmr == 3: set_message("........")
            if tmr == 4: set_message("...........")
            if tmr == 5:
                if random.randint(0, 99) < 60:
                    idx = 22
                else:
                    set_message("You failed to flee")
            if tmr == 10:
                idx = 13
                tmr = 0
        # 패배
        elif idx == 15:
            draw_battle(screen, fontS)
            if tmr == 1:
                pygame.mixer.music.stop()
                set_message("You lose")
            if tmr == 11:
                idx = 9
                tmr = 29
                
        # 승리
        elif idx == 16:
            draw_battle(screen, fontS)
            if tmr == 1:
                set_message("You win!")
                pygame.mixer.music.stop()
                se[5].play()
            if tmr == 28:
                idx = 22
                if random.randint(0, emy_lifemax) > random.randint(0, pl_lifemax):
                    idx = 17
                    tmr = 0
                    
        # 레벨 업 
        elif idx == 17:
            draw_battle(screen, fontS)
            if tmr == 1:
                set_message("Level up!")
                se[4].play()
                lif_p = random.randint(10, 20)
                str_p = random.randint(5, 10)
            if tmr == 21:
                set_message("Max life + " + str(lif_p))
                pl_lifemax = pl_lifemax + lif_p
            if tmr == 26:
                set_message("Str + " + str(str_p))
                pl_str += str_p
            if tmr == 50:
                idx = 22
                
        # Potion
        elif idx == 20:
            draw_battle(screen, fontS)
            if tmr == 1:
                set_message("Potion!")
                se[2].play()
            if tmr == 6:
                pl_life = pl_lifemax
                potion -= 1
            if tmr == 11:
                idx = 13
                tmr = 0
                
        # Blaze gem
        elif idx == 21:
            draw_battle(screen, fontS)
            img_rz = pygame.transform.rotozoom(imgEffect[1], 30 * tmr, (12 - tmr) / 8)
            X = 440 - img_rz.get_width() / 2
            Y = 360 - img_rz.get_height() / 2
            screen.blit(img_rz, [X, Y])
            if tmr == 1:
                set_message("Blaze gem!")
                se[1].play()
            if tmr == 6:
                blazegem -= 1
            if tmr == 11:
                dmg = 1000
                idx = 12
                tmr = 4
                
        # 전투 종료
        elif idx == 22:
            pygame.mixer.music.load("sound/ohd_bgm_field.ogg")
            pygame.mixer.music.play(-1)
            idx = 1
            tmr = 11
            
        elif idx == 23:
            draw_dungeon(screen, fontS)
            if 1 <= tmr and tmr <= 10:
                draw_text(screen, "Click T again to return to title", 320, 560, fontS, WHITE)
                if key[K_t] == 1:
                    idx = 0
                    tmr = 0
                    attack_type = 0
            if tmr == 11:
                idx = 1
                tmr = 0
        # 엔딩
        elif idx == 24:
            draw_text(screen, "End", 420, 360, fontS, WHITE)
            draw_text(screen, "Press SPACE to return to title", 295, 400, fontS, WHITE)
            if key[K_SPACE] == 1:
                idx = 0
                tmr = 0
        
        # intro
        elif idx == 25:
            # video.preview()
            idx = 0
            tmr = 0
            
        if idx != 24:
            draw_text(screen, "[S]peed" + str(speed), 740, 40, fontS, WHITE)
            
        pygame.display.update()
        clock.tick(30)

if __name__ == '__main__':
    main()
```
